# 1 Generating the images locally

## 1 Imports

```python
import numpy as np
import pandas as pd
import ast
import os
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette(sns.color_palette('Blues_r', 20))
```

The code cell below will take 1000 drawing values of each existing category and store these locally. 

This process took 2 minutes 53 seconds and the 5000 images only take up 59mb of space.

## 2 Drawing and saving the images

```python
directory = '../Data/10_categories/'
images = '../Data/Images/'
from matplotlib.pyplot import figure
# loop over the directory containing the csvs of categories
for file in os.listdir(directory):
    count = 0

    filename = file.split('.')[0]

    # check if folder already exists, if not make it
    if not os.path.exists(images+str(filename)):
        os.makedirs(images+str(filename))

    df = pd.read_csv(str(directory)+str(file))
    df = df.sample(n=10, random_state=1).reset_index(drop=True)
    df['drawing'] = df['drawing'].apply(ast.literal_eval)

    # plot the strokes and save the plot as an image
    for i, drawing in enumerate(df.drawing):
        count = count + 1

        for x, y in drawing:
            plt.plot(x, -np.array(y), lw=3)
            plt.axis('off')

        plt.axis('square')
        plt.show()
```

## 3 Resizing the images

Resizing the images to a width of 256 while maintaining the proper aspect ratio: all images have a height of 170 in this instance.

```python
from PIL import Image
count = 0
animals = ['airplane','alarm clock','ambulance','angel','animal migration','ant','anvil','apple','arm','asparagus']

# loop over every animal folder
for i in range(len(animals)):
  print(animals[i])

  # loop over every animal drawing in the current animal folder
  for a, filename in enumerate(os.listdir('../Data/Images/'+animals[i])):
      size = 256
      img = Image.open("../Data/Images/"+animals[i]+"/"+filename)
      img = img.resize((size,size), Image.ANTIALIAS)
      img.save("../Data/Images/"+animals[i]+"/"+filename)
      count += 1
      print(count)
```

Checking if all images have a width of 256

```python
import PIL
error = 0
animals = ['butterfly','cat','dog','elephant','bee']

for i in range(len(animals)):
  for a, filename in enumerate(os.listdir("Images/"+animals[i])):
    image = PIL.Image.open("Images/"+animals[i]+"/"+filename)
    width, height = image.size
    if not width == 256:
      error += 1
      print(animals[i]+"/"+filename+"has "+width+" width")

print("Images with width different from 256: "+str(error))
```
