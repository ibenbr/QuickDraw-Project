# Visualising the images

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast
import json
sns.set_palette(sns.color_palette('copper', 20))
```

```python
import ast
import pandas as pd
df = pd.read_csv('5_animals.csv')[-100:]
df['drawing'] = df['drawing'].apply(ast.literal_eval)
```

```python
n = 10
fig, axs = plt.subplots(nrows=n, ncols=n, sharex=True, sharey=True, figsize=(16, 10))
for i, drawing in enumerate(df.drawing):
    ax = axs[i // n, i % n]
    print(i)
    print(drawing)
    for x, y in drawing:
        ax.plot(x, -np.array(y), lw=3)
    ax.axis('off')

#uncomment to save the plot as an image
#fig.savefig('animals.png', dpi=200) 

plt.show();
```
