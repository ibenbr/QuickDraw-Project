# example of how the drawing column works

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
df = pd.read_csv('5_animals.csv')
```

```python
for index, row in df.head(5).iterrows():    
    drawing = df['drawing'].iloc[index]
    drawing = json.loads(drawing)
    print(str(index)+": "+df['word'].iloc[index])
    plt.figure()

    for x,y in drawing:
        plt.plot(x,y)
    plt.tight_layout()
    plt.gca().invert_yaxis()
    #fig.savefig('gs.png', dpi=300)
    plt.show()
```
