```python
import numpy as np
import pandas as pd
import tensorflow as tf
import cv2
import ast
import json
import os
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette(sns.color_palette('copper', 20))
```

```python
directory = '5_animals/'
images = 'Images/'

for file in os.listdir(directory):
    print(file)
    filename = file.split('.')[0]
    if not os.path.exists(images+str(filename)):
        os.makedirs(images+str(filename))

    df = pd.read_csv(str(directory)+str(file))
    df = df[df["recognized"]==True]
    df = df.drop(columns=['countrycode','recognized','timestamp','word','key_id']).reset_index(drop=True)
    
    df = df.sample(n=1000, random_state=1).reset_index(drop=True)
    df['drawing'] = df['drawing'].apply(ast.literal_eval)

    count = 0
    for i, drawing in enumerate(df.drawing):
        count = count + 1
        print(str(count)+': '+str(filename))
        for x, y in drawing:
            plt.plot(x, -np.array(y), lw=3)
            plt.axis('off')

        plt.savefig(images+str(filename)+'/'+str(i)+'.png')
        plt.close()

```
